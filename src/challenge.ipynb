# Ejercicio 1: 
Las top 10 fechas donde hay más tweets. Mencionar el usuario (username) que más publicaciones tiene por cada uno de esos días. 

**Supuesto1:** Para cumplir con lo solicitado, se trabaja el USERNAME como id unico representante de una cuenta, pero, como regla de negocio esto es incorrecto, ya que un USERNAME en twitter puede cambiar para una misma cuenta.




```python
%load_ext memory_profiler

from q1_time   import q1_time   as chlatam_q1t
from q1_memory import q1_memory as chlatam_q1m
from q2_time   import q2_time   as chlatam_q2t
from q2_memory import q2_memory as chlatam_q2m
from q3_time   import q3_time   as chlatam_q3t
from q3_memory import q3_memory as chlatam_q3m
from memory_profiler import memory_usage,profile
import cProfile,pstats

str_input='../../farmers-protest-tweets-2021-2-4.json'
```

## Ejecucion1.1 optimizada para Tiempo ##

```python
%reload_ext memory_profiler

list_chlatam=[]

def resp_ch():
    global list_chlatam
    list_chlatam=chlatam_q1t(str_input) ###########CALL

def try_stat():
    global list_chlatam
    profiler=cProfile.Profile()
    profiler.enable()
    memory_usage_result = memory_usage(resp_ch,max_usage=True)
    profiler.disable()
    
    print('RESUME:------------------------------------------------------------------------------------')
    print('         Maxima de memoria utilizada: '+str(round(memory_usage_result,3))+' MB')
    stats = pstats.Stats(profiler)
    stats.print_stats(0)
    print('RETURN:------------------------------------------------------------------------------------')
    for item in list_chlatam:
        print('         '+str(item))

if __name__ == '__main__':
    try_stat()

```
